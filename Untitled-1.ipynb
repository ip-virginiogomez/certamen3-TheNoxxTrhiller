```python
#Control de velocidad en autopistas.

velocidades = []

print("Ingrese las 5 velocidades:")
for i in range(5):
    v = float(input("Velocidad: "))
    velocidades.append(v)

suma = 0
maxima = velocidades[0] 

for v in velocidades:
    suma = suma + v        
    if v > maxima:
        maxima = v         

promedio = suma / 5

print("\nResultado")
print("Promedio:", promedio, "km/h")
print("Maxima:", maxima, "km/h")

print("\n Alertas")

todos_bien = True
for v in velocidades:
    if v < 60 or v > 120:
        todos_bien = False

if todos_bien == False:
    print("Al menos un vehiculo no cumplio con el limite (60-120 km/h).")
else:
    print("Todos los vehiculos estan dentro del limite permitido.")

for v in velocidades:
    if v < 20 or v > 140:
        print("ADVERTENCIA DE PELIGRO: Se detectaron velocidades extremas")
        break
```

```python
#Registro de Ventas de una Tienda

matriz_ventas = [
    [20000, 15000, 10000],
    [5000, 10000, 8000],
    [30000, 25000, 40000]
]

total_v1 = matriz_ventas[0][0] + matriz_ventas[0][1] + matriz_ventas[0][2]
total_v2 = matriz_ventas[1][0] + matriz_ventas[1][1] + matriz_ventas[1][2]
total_v3 = matriz_ventas[2][0] + matriz_ventas[2][1] + matriz_ventas[2][2]

print("Totales por Vendedor")
print("Julian:", total_v1)
print("Pepito:", total_v2)
print("patricio balboa:", total_v3)

print("\n Mejor Rendimiento")
if total_v1 > total_v2 and total_v1 > total_v3:
    print("El vendedor con mayor rendimiento fue Julian")
elif total_v2 > total_v1 and total_v2 > total_v3:
    print("El vendedor con mayor rendimiento fue Pepito")
else:
    print("El vendedor con mayor rendimiento fue Patricio Balboa")

print("\n Alertas de bajo desempeno")
if total_v1 < 30000:
    print("Alerta: El Vendedor 1 tuvo ventas menores a $30.000.")

if total_v2 < 30000:
    print("Alerta: El Vendedor 2 tuvo ventas menores a $30.000.")

if total_v3 < 30000:
    print("Alerta: El vendedor 3 tuvo ventas menores a $30.000.")
```

```python
#Sistema de Descuento en Supermercado

edad = int(input("¿Cuántos años tiene el cliente?: "))
tarjeta = input("¿Tiene tarjeta de socio? (si/no): ")
compra = float(input("¿Cuánto es el total de la compra?: $"))

if compra > 10000 and (edad > 60 or tarjeta == "si"):
    
    descuento = compra * 0.15
    total_final = compra - descuento
    print("¡Obtiene el 15% de descuento!")
    print("Monto final a pagar: $", total_final)

else:
    
    print("No califica para el descuento.")
    print("Monto final a pagar: $", compra)
```
